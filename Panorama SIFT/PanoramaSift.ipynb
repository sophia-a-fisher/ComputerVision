```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imPath = '../Images/mtn1.png'
imPath2 = '../Images/mtn2.png'
img1 = cv2.imread(imPath);
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread(imPath2);
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.SIFT_create()
kp = sift.detect(gray1,None)
kp2 = sift.detect(gray2,None)
```

```python
img = cv2.drawKeypoints(gray1,kp,img1,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img2 = cv2.drawKeypoints(gray2,kp2,img2,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(img[:,:,::-1])
```

```python
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(img2[:,:,::-1])
```

```python
img1 = cv2.imread(imPath);
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread(imPath2);
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
```

```python
matches = bf.match(des1, des2)
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:20], img2, flags=2)
```

```python
plt.figure(figsize=(15,20))
plt.imshow(img3[:,:,::-1])
```
